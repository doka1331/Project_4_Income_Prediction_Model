```python
# Import dependencies
import pandas as pd
import numpy as np
# Read the data into a Pandas DataFrame
US_adult_income_df = pd.read_csv('Resources/adult.csv')
US_adult_income_df.head()
```

```python
#Brief Summary of Raw Data
US_adult_income_df.info()
```

```python
#Dropping columns "fnlwgt, education.num, capital.gain, capital.loss, native.country" from US_adult_income_df
US_adult_income_df = US_adult_income_df.drop(columns = ["fnlwgt", "education.num", "capital.gain", "capital.loss", "native.country"])

# Display the dataset to verify the columns have been dropped
US_adult_income_df
```

```python
#Dropping any row with missing values
# Replace "?" with NaN
US_adult_income_df.replace("?", np.nan, inplace=True)

# Drop rows with any NaN values
US_adult_income_df.dropna(inplace=True)

#Display the dataset to verify the rows have been dropped
US_adult_income_df
```

```python
#Replace the 'income' column values of '<=50k'and '>50k' with '0' and '1' values
US_adult_income_df['income'] = US_adult_income_df['income'].replace({'<=50K': 0, '>50K': 1})

#Rename the 'income' column to 'Income over 50k? 0=no 1=yes'
US_adult_income_df.rename(columns={'income': 'Income over 50k? 0=no 1=yes'}, inplace=True)

#Display the dataset to verify the rows have been dropped
US_adult_income_df
```

```python
#Copy the cleaned dataset to a new dataset name 'US_adult_income_cleaned'
US_adult_income_cleaned = US_adult_income_df.copy()
#display this new dataset
US_adult_income_cleaned
```

```python
#copy new dataset to new output csv file named 'adult_cleaned.csv'
import os
# Define the path to the resources folder
resources_folder = 'Resources'
os.makedirs(resources_folder, exist_ok=True)  # Create the folder if it doesn't exist

# Define the full path for the new CSV file
csv_file_path = os.path.join(resources_folder, 'adult_cleaned.csv')

# Save the cleaned dataset to the new CSV file
US_adult_income_cleaned.to_csv(csv_file_path, index=False)
```
